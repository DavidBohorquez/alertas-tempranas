```python
df['Target'][df['Target']=='Enrolled'].count()

df.drop(df[df['Target']=='Enrolled'].index, inplace=True)

df.shape
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

aux_df = df.copy()

aux_df['Target'] = label_encoder.fit_transform(aux_df['Target'])

aux_df['Target'].unique()
df['Target'].unique()

def norm(item):
  return (item)/(item.abs().max())

aux_df = aux_df.apply(norm)
aux_df
```

```python
import matplotlib.pyplot as plt

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.preprocessing import StandardScaler

X = aux_df.iloc[:,0:36]
y = aux_df.iloc[:,36:37]

model = ExtraTreesClassifier()
model.fit(X,y)

print(model.feature_importances_)

feat_importance = pd.Series(model.feature_importances_, index=X.columns)
feat_importance.nlargest(20).plot(kind='barh')
plt.show()
```

```python
import numpy as np

filtered_df = df.iloc[:,np.r_[1,3,6,8,11:13,15:20,22:26,28:32,33,-1]]

filtered_df.columns
filtered_df.shape
```

```python
label_encoder = LabelEncoder()

filtered_df['Target'] = label_encoder.fit_transform(filtered_df['Target'])
```

```python
from sklearn.model_selection import train_test_split

X = filtered_df.iloc[:,0:-1]
y = filtered_df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
features_train_scale = scaler.transform(X_train)
features_test_scale = scaler.transform(X_test)
X_train
features_train_scale

from sklearn.neural_network import MLPClassifier
iterations=1000
hidden_layers=[10,10,10]

mlp = MLPClassifier(hidden_layer_sizes=(hidden_layers), max_iter=iterations) 

mlp.fit(X_train, y_train)
```

```python
predicted = mlp.predict(X_test)

print('Predected value\n',predicted,'\n')
print('Real value\n',y_test.values.ravel(),'\n')
print('Score:',mlp.score(X_test, y_test))

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
print("\nEvaluation: considering the confusion matrix")
print(confusion_matrix(y_test,predicted), '\n')  

cm = confusion_matrix(y_test, predicted, labels=mlp.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=mlp.classes_)
disp.plot()
plt.show()
```

```python
import pickle

pickle.dump(mlp, open('Hack4Edu.pkl', 'wb'))
```
